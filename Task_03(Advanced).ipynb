```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
dt=pd.read_csv('spotify_data.csv')
```

```python
dt.head()
```

```python
dt.info()
```

```python
dt.drop_duplicates(inplace=True)
```

```python
dt.info()
dt.shape
```

```python
dt.describe()
```

```python
col=dt.columns
col
```

```python
col=['popularity',
       'danceability', 'energy', 'loudness', 'speechiness',
       'acousticness', 'liveness', 'valence', 'tempo']
```

```python
fig, ax = plt.subplots(3,3, figsize=(10,10))
plt.title('Feature Distribution')
i=0
j=0
for feature in col:
    ax[i,j].hist(dt[feature])
    ax[i,j].set_title(feature)
    j+=1
    if j>2:
        j=0
        i+=1
    if i>2:
        break
plt.show()
```

```python
genre_dist=dt[['genre','song']]
genre_dist=genre_dist.groupby('genre',as_index=False).count().sort_values(by='song',ascending=False).head(25)
genre_dist.head()
```

```python
plt.figure(figsize=(20,5))
plt.title('No. of song in each Genre')
plt.xlabel('Genre')
plt.ylabel('No. of songs')
plt.bar(genre_dist['genre'],genre_dist['song'])
plt.xticks(rotation=-90)
plt.show()
```

```python
year_dist=dt[['year','song']]
year_dist=year_dist.groupby('year',as_index=False).count().sort_values(by='song',ascending=False).sort_values(by='year',ascending=True)
year_dist.head()
```

```python
plt.plot(year_dist['year'],year_dist['song'],marker='.',color='black')
plt.fill_between(year_dist['year'],year_dist['song'],alpha=0.75,color='skyblue')
plt.title('No. of Songs per Year')
plt.xlabel('Year')
plt.ylabel('No. of Songs')
plt.show()
```

```python
popularity_dist=dt[['genre','popularity']]
popularity_dist=popularity_dist.groupby('genre',as_index=False).sum().sort_values(by='popularity',ascending=False).head(30)
plt.figure(figsize=(15,5))
plt.bar(popularity_dist['genre'],popularity_dist['popularity'])
plt.title('Genres by Popularity')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.xticks(rotation=-90)
plt.show()
```

```python
top_artist=dt[['artist','song']]
top_artist=top_artist.groupby('artist',as_index=False).count().sort_values(by='song',ascending=False).head(50)
```

```python
plt.figure(figsize=(12,4))
plt.bar(top_artist['artist'],top_artist['song'])
plt.title('No. of songs by Artist')
plt.xlabel('Artist')
plt.ylabel('Sonng')
plt.xticks(rotation=-90)
plt.show()
```

```python
popu_artist=dt[['artist','popularity']]
popu_artist=popu_artist.groupby('artist',as_index=False).sum().sort_values(by='popularity',ascending=False).head(25)
```

```python
plt.figure(figsize=(10,5))
plt.bar(popu_artist['artist'],popu_artist['popularity'])
plt.title('Top 25 Artist by Popularity')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=-90)
plt.show()
```

```python
top_songs=dt[['song','popularity']]
top_songs=top_songs.sort_values(by='popularity',ascending=False).head(30)
```

```python
plt.figure(figsize=(10,5))
plt.plot(top_songs['song'],top_songs['popularity'],color='g',marker='.')
plt.title('Top 30 Songs by Popularity')
plt.xlabel('Songs')
plt.ylabel('Popularity')
plt.xticks(rotation=-90)
plt.show()
```

```python
corr=dt.drop(['artist','song','genre'],axis=1).corr()
fig, ax = plt.subplots(figsize=(5,5))
ticks=np.arange(len(corr.columns))
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
plt.title('Correlation Matrix')
sns.heatmap(corr,cmap='viridis')
plt.xticks(rotation=-90)
plt.show()
```

```python
key_dist=dt[['key','popularity']]
key_dist=key_dist.groupby('key',as_index=False).sum().sort_values(by='popularity').head(7)
key_dist
```

```python
plt.title('Top 7 Key based on Popularity')
plt.pie(key_dist['popularity'],labels=key_dist['key'],startangle=90,autopct='%1.1f%%',colors=sns.color_palette('colorblind'))
plt.show()
```

```python
mode_dist=dt[['mode','popularity']]
mode_dist=mode_dist.groupby('mode',as_index=False).sum().sort_values(by='popularity')
mode_dist
```

```python
plt.title('Popularity of Mode')
plt.pie(mode_dist['popularity'],labels=mode_dist['mode'],startangle=90,autopct='%1.1f%%',colors=['Red','Grey'])
plt.show()
```
